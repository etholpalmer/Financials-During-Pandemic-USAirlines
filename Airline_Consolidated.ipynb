## **Gathering the fundamental data for the stocks**

The Quarterly Financial Statement data is gathered from 
* **SIMFIN API** for the period from 2011 to 2020
* StockAnalysis Website (https://stockanalysis.com/) for the year 2020 and 2021

```python
# Using SimFin API to get the financial statement data from 2011 to 2020
import pandas as pd
import simfin as sf
import numpy as np
%matplotlib inline
```

```python
sf.set_data_dir('~/Data/')      # Changed from '~/Data/' Data in the home directory
```

```python
sf.set_api_key(api_key='free')
```

```python
%%time
market = 'us'
df_income = sf.load_income(variant='quarterly', market=market)
df_bal_sheet =sf.load_balance(variant='quarterly', market=market)
```

```python
#Filtering the fundamental data for stocks of interest
tickers = ["LUV","DAL","UAL"]
airline_income_df=df_income.loc[tickers].reset_index()
airline_balsheet_df=df_bal_sheet.loc[tickers].reset_index()
```

```python
airline_income_df.tail()
airline_income_df.columns
```

```python
# Dropping the columns from the data frame
drop_inc_columns=['SimFinId','Currency','Restated Date','Shares (Diluted)','Selling, General & Administrative','Research & Development', 'Depreciation & Amortization','Non-Operating Income (Loss)',
        'Pretax Income (Loss), Adj.','Abnormal Gains (Losses)','Income Tax (Expense) Benefit, Net','Income (Loss) from Continuing Operations',
        'Net Extraordinary Gains (Losses)','Net Income (Common)']
airline_income_df.drop(columns=drop_inc_columns,inplace=True)
```

```python
airline_income_df.head()
```

```python
# Reading the excel file for income statement data for 2020 and 2021
income_df=pd.read_excel("Data/Airline_income.xlsx")
income_df.head()
```

```python
airline_income_df=pd.concat([airline_income_df,income_df],axis='rows')
```

```python
#airline_income_df.loc[airline_income_df['Ticker']=='UAL'].tail()
airline_income_df.head()
#airline_balsheet_df.columns
```

```python
#drop columns -SimFinid, currency, Restated Date,Shares(Diluted),Share Capital & Additional Paid-In Capital', 'Treasury Stock' 'Retained Earnings'
drop_bal_columns=['SimFinId', 'Currency','Restated Date','Shares (Diluted)','Shares (Basic)','Share Capital & Additional Paid-In Capital','Treasury Stock','Retained Earnings']
airline_balsheet_df.drop(columns=drop_bal_columns,inplace=True)
airline_balsheet_df.head()
```

```python
# Reading the excel file for income statement data for 2020 and 2021
balsheet_df=pd.read_excel("Data/Airline_balsheet.xlsx")
balsheet_df.head()
```

```python
#balsheet_df
airline_balsheet_df=pd.concat([airline_balsheet_df,balsheet_df],axis='rows')
```

```python
#airline_balsheet_df.loc[airline_balsheet_df['Ticker']=='DAL'].tail()
airline_balsheet_df.head()
```

```python
#Calculating the Balance Sheet Ratios

airline_balsheet_df['Quick Ratio']=(airline_balsheet_df['Total Current Assets']-airline_balsheet_df['Inventories'])/airline_balsheet_df['Total Current Liabilities']
airline_balsheet_df['Current Ratio']=airline_balsheet_df['Total Current Assets']/airline_balsheet_df['Total Current Liabilities']
airline_balsheet_df['Debt-to-Equity Ratio']=(airline_balsheet_df['Long Term Debt']+ airline_balsheet_df['Short Term Debt'])/airline_balsheet_df['Total Equity']
airline_balsheet_df['Long Term Debt-to-Equity Ratio']=airline_balsheet_df['Long Term Debt']/airline_balsheet_df['Total Equity']
airline_balsheet_df['Short Term Debt-to-Equity Ratio']=airline_balsheet_df['Short Term Debt']/airline_balsheet_df['Total Equity']
```

```python
#calculating income statement ratios

airline_income_df["Gross Profit Margin"]=airline_income_df['Gross Profit']/airline_income_df['Revenue']
airline_income_df["Net Profit Margin"]=airline_income_df['Net Income']/airline_income_df['Revenue']
airline_income_df["Operating Margin"]=airline_income_df['Operating Income (Loss)']/airline_income_df['Revenue']
airline_income_df['EPS']=airline_income_df['Net Income']/airline_income_df['Shares (Basic)']
airline_income_df['Interest Coverage Ratio']=airline_income_df['Operating Income (Loss)']/airline_income_df['Interest Expense, Net']
```

```python
inc_columns=['Ticker','Report Date','Fiscal Year','Fiscal Period','Shares (Basic)',
             'Revenue','Net Income','Gross Profit Margin','Net Profit Margin',
             'Operating Margin', 'EPS','Interest Coverage Ratio']
bal_columns=['Cash, Cash Equivalents & Short Term Investments','Total Assets','Total Equity',
             'Quick Ratio', 'Current Ratio', 'Debt-to-Equity Ratio','Long Term Debt-to-Equity Ratio',
             'Short Term Debt-to-Equity Ratio']
```

```python
# Joining the Balance sheet data and Income Statement Data
LUV_balsheet=airline_balsheet_df.loc[airline_balsheet_df['Ticker']=='LUV']
LUV_balsheet.set_index('Publish Date',inplace=True)
LUV_income=airline_income_df.loc[airline_income_df['Ticker']=='LUV']
LUV_income.set_index('Publish Date',inplace=True)
LUV_fundamentals=pd.concat([LUV_income[inc_columns],LUV_balsheet[bal_columns]],join='inner',axis='columns')
LUV_fundamentals['ROE']=LUV_fundamentals['Net Income']/LUV_fundamentals['Total Equity']
```

```python
DAL_balsheet=airline_balsheet_df.loc[airline_balsheet_df['Ticker']=='DAL']
DAL_balsheet.set_index('Publish Date',inplace=True)
DAL_income=airline_income_df.loc[airline_income_df['Ticker']=='DAL']
DAL_income.set_index('Publish Date',inplace=True)
DAL_fundamentals=pd.concat([DAL_income[inc_columns],DAL_balsheet[bal_columns]],join='inner',axis='columns')
DAL_fundamentals['ROE']=DAL_fundamentals['Net Income']/DAL_fundamentals['Total Equity']
```

```python
UAL_balsheet=airline_balsheet_df.loc[airline_balsheet_df['Ticker']=='UAL']
UAL_balsheet.set_index('Publish Date',inplace=True)
UAL_income=airline_income_df.loc[airline_income_df['Ticker']=='UAL']
UAL_income.set_index('Publish Date',inplace=True)
UAL_fundamentals=pd.concat([UAL_income[inc_columns],UAL_balsheet[bal_columns]],join='inner',axis='columns')
UAL_fundamentals['ROE']=UAL_fundamentals['Net Income']/UAL_fundamentals['Total Equity']
```

## **Combing the stock price data with Fundamental data to calculate valuation ratios**

```python
# Filtering the airline stock price data for closing prices
LUV_stock=airline_stock['LUV']['close']
DAL_stock=airline_stock['DAL']['close']
UAL_stock=airline_stock['UAL']['close']
```

```python
# Concatenating the LUV stock prices to the ratios
LUV_data=pd.concat([LUV_fundamentals,LUV_stock],axis='columns',join='inner')
LUV_data.rename(columns={"close":"LUv Price"},inplace=True)

#Calculating P/E ,P/S and P/B ratios for Southwest Airlines
LUV_data["PE Ratio"]=LUV_data["LUv Price"]/LUV_data["EPS"]
LUV_data["PS Ratio"]=(LUV_data["LUv Price"]*LUV_data["Shares (Basic)"])/LUV_data["Revenue"]
LUV_data["PB Ratio"]=(LUV_data["LUv Price"]*LUV_data["Shares (Basic)"])/LUV_data["Total Equity"]
```

```python
#DAL_data.shape
```

```python
# Concatenating the DAL stock prices to the ratios
DAL_data=pd.concat([DAL_fundamentals,DAL_stock],axis='columns',join='inner')
DAL_data.rename(columns={"close":"DAL Price"},inplace=True)

#Calculating P/E ,P/S and P/B ratios for Delta Airlines
DAL_data["PE Ratio"]=DAL_data["DAL Price"]/DAL_data["EPS"]
DAL_data["PS Ratio"]=(DAL_data["DAL Price"]*DAL_data["Shares (Basic)"])/DAL_data["Revenue"]
DAL_data["PB Ratio"]=(DAL_data["DAL Price"]*DAL_data["Shares (Basic)"])/DAL_data["Total Equity"]
```

```python
# Concatenating the UAL stock prices to the ratios
UAL_data=pd.concat([UAL_fundamentals,UAL_stock],axis='columns',join='inner')
UAL_data.rename(columns={"close":"UAL Price"},inplace=True)

#Calculating P/E ,P/S and P/B ratios for United Airlines
UAL_data["PE Ratio"]=UAL_data["UAL Price"]/UAL_data["EPS"]
UAL_data["PS Ratio"]=(UAL_data["UAL Price"]*UAL_data["Shares (Basic)"])/UAL_data["Revenue"]
UAL_data["PB Ratio"]=(UAL_data["UAL Price"]*UAL_data["Shares (Basic)"])/UAL_data["Total Equity"]
```
