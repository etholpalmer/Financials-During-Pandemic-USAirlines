```python
# Created a create_bar_chart function

def create_bar_chart(data, title, xlabel, ylabel, color=None, height=500, width=500, rot=60):
    
    from bokeh.models.formatters import BasicTickFormatter

    hv.extension('bokeh')
    if color is None:
      return data.hvplot.bar(
          stacked=False
          , title=title
          , height=height
          , width=width
          , rot=rot
          , xlabel=xlabel
          , ylabel=ylabel
          , yformatter=BasicTickFormatter(use_scientific=False)
      )
    elif (color is not None):
      return data.hvplot.bar(
          stacked=False
          , title=title
          , height=height
          , width=width
          , rot=rot
          , xlabel=xlabel
          , ylabel=ylabel
          , yformatter=BasicTickFormatter(use_scientific=False)
      ).opts(color=color)
```
