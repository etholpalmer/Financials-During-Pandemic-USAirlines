```python
import os
import requests
from dotenv import load_dotenv
import pandas as pd
import csv as csv
import alpaca_trade_api as tradeapi
```

```python
load_dotenv()
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("alpaca_akey")
alpaca_secret_key = os.getenv("alpaca_skey")
type(alpaca_secret_key)
```

```python
# Create the Alpaca API object
api=tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2"
)
```

```python
tickers=["LUV","DAL","UAL","AAL"]
timeframe="1D"
```

```python
start_date1=pd.Timestamp("2011-01-01",tz="America/New_York").isoformat()
end_date1=pd.Timestamp("2014-01-01",tz="America/New_York").isoformat()
```

```python
df_stock_data1=api.get_barset(
        tickers,
        timeframe,
        start=start_date1,
        end=end_date1,
        limit=1000

).df
```

```python
df_stock_data1.tail()
```

```python
start_date1=pd.Timestamp("2014-01-01",tz="America/New_York").isoformat()
end_date1=pd.Timestamp("2018-01-01",tz="America/New_York").isoformat()

df_stock_data2=api.get_barset(
        tickers,
        timeframe,
        start=start_date1,
        end=end_date1,
        limit=1000

).df
```

```python
df_stock_data2.tail()
```

```python
start_date1=pd.Timestamp("2018-01-01",tz="America/New_York").isoformat()
end_date1=pd.Timestamp("2021-06-25",tz="America/New_York").isoformat()

df_stock_data3=api.get_barset(
        tickers,
        timeframe,
        start=start_date1,
        end=end_date1,
        limit=1000

).df
```

```python
df_stock_data3.tail()
```

```python
airline_stock=pd.concat([df_stock_data1,df_stock_data2,df_stock_data3])
```
